```python
#change format of timestamp for time series analysis

renamed_col['TimeStamp'] = pd.to_datetime(renamed_col['TimeStamp'], format='%d/%m/%Y %H:%M')

#sort values for sequential processing

renamed_col.sort_values(by=['Run', 'TimeStamp'], inplace=True)
```

```python
def make_timestamps_consecutive(group):
    # Ensure the group is sorted (should be already from previous steps)
    group = group.sort_values(by='TimeStamp')

    # Detect consecutive duplicates and increment
    while group['TimeStamp'].duplicated().any():
        group['TimeStamp'] += pd.to_timedelta(group.groupby('TimeStamp').cumcount(), unit='s')

    return group

# Apply this function to each group defined by 'Run'
apply_second_offset = renamed_col.groupby('Run').apply(make_timestamps_consecutive)
```

```python
# Optionally reset index if it was altered during processing
apply_second_offset.reset_index(drop=True, inplace=True)
```

```python
# Renaming columns without converting to lowercase
apply_second_offset.rename(columns=lambda x: x.strip().replace(' ', '_'), inplace=True)
```

```python
# Function to find columns with identical values
def find_identical_columns(df):
    identical_columns = {}  # Dictionary to store pairs of identical columns
    columns = df.columns
    for i in range(len(columns)):
        for j in range(i + 1, len(columns)):
            if df[columns[i]].equals(df[columns[j]]):
                if columns[i] not in identical_columns:
                    identical_columns[columns[i]] = []
                identical_columns[columns[i]].append(columns[j])
    return identical_columns

# Apply the function to your DataFrame
identical_cols = find_identical_columns(apply_second_offset)
print(identical_cols)
```

```python
from sklearn.preprocessing import OneHotEncoder


# Assume you've identified categorical columns. Let's say 'Run' and 'MachineRecipeID' are categorical.
categorical_features = ['LOT_ID',	'MachineRecipeID',	'PortID',	'ProcessOpNum',	'ProductGrpID',	'ProductID',	'RECIPE_ID',	'RouteID',	'Technology']

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse=False)  # sparse=False means the output will be a dense matrix

# Fit and transform the categorical features
encoded_features = encoder.fit_transform(drop_col[categorical_features])

# Create a DataFrame with the encoded features, note the feature names can be accessed via encoder.get_feature_names()
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names(categorical_features))

# Concatenate the original DataFrame with the new encoded DataFrame
encoded_data = pd.concat([drop_col, encoded_df], axis=1)

# Optionally, drop the original categorical columns if they are no longer needed
encoded_data.drop(columns=categorical_features, inplace=True)

# Display the first few rows of the updated DataFrame to verify
display(encoded_data.head())
```
