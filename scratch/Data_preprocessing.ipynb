```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
from IPython.display import display

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', None)

file_path = 'Tool_Sensor_Data.csv'
sensor_data = pd.read_csv(file_path)
sensor_data.head() 
sensor_data.info() 
sensor_data.describe()
```

```python
print(sensor_data.isnull().sum())
```

```python
threshold = len(sensor_data) * 0.5
sensor_data = sensor_data.dropna(thresh=threshold, axis=1)
print("\nColumns with excessive missing values removed. Remaining columns:")
display(sensor_data.columns)
```

```python
print("\nHandling remaining missing values...")
sensor_data.fillna(sensor_data.mean(), inplace=True)
print("Missing values handled. Current missing values count:")
display(sensor_data.isnull().sum())
```
