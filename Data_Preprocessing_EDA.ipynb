### Remove rows and columns with missing values

```python
# Identify columns that are completely empty
empty_columns = sensor_data.columns[sensor_data.isnull().all()]

# Drop these columns from the DataFrame
removed_empty_column = sensor_data.drop(columns=empty_columns)

print(f"Removed empty columns: {empty_columns}")
```

```python
# Verify if columns have been removed
removed_empty_column.head()
```

```python
# Remove rows with majority of missing values
# Calculate the threshold for 'mostly missing' as more than 50% missing values in a row
threshold = len(removed_empty_column.columns) * 0.5

# Identify rows that have more than 50% of their values missing
rows_with_mostly_missing_values = removed_empty_column[removed_empty_column.isnull().sum(axis=1) > threshold]


# Remove these rows from the main DataFrame
removed_empty_rows = removed_empty_column.drop(rows_with_mostly_missing_values.index)
```

```python
# Drop columns that are empty
# List of columns to drop
columns_to_drop = ['EventType', 'EventName', 'EventId']

# Drop the empty columns
drop_specific_col = removed_empty_rows.drop(columns=columns_to_drop)

# Display the DataFrame to confirm the columns have been removed
display(drop_specific_col.head())
```

```python
# Detect columns with constant values by checking if the number of unique values is 1
constant_columns = [col for col in drop_specific_col.columns if drop_specific_col[col].nunique() <= 1]

# Drop these constant columns
drop_const_col = drop_specific_col.drop(columns=constant_columns)

# Display the names of the removed columns and the updated DataFrame to verify
print("Removed constant columns:", constant_columns)
display(drop_const_col.head())
```

```python
# Specify new names for sensor reading columns

# rename from random names to 'Sensor_1', 'Sensor_2', 'Sensor_3' and so on
start_index = list(drop_const_col.columns).index('OunhHslCRwIRilo')  # Get the starting index
end_index = start_index + 33  # Define the end index based on how many columns to rename

# Generate new names on the specific range of columns
new_names = ['Sensor_' + str(i + 1) for i in range(end_index - start_index)]
drop_const_col.columns.values[start_index:end_index] = new_names
```

```python
# Display the updated sensor columns
display(drop_const_col.head())
```

```python
# rename dataframe 
renamed_col = drop_const_col

# backup dataframe
backup_data_after_col_rename = renamed_col.copy()

# Save the DataFrame to a CSV file
backup_data_after_col_rename.to_csv('initial_data.csv', index=False)
```

```python
display(renamed_col.head())
```

```python
# Check for leading/trailing whitespace in column names
renamed_col.columns = renamed_col.columns.str.strip()

# Identify columns with identical values
# Function to find columns with identical values
def find_identical_columns(df):
    identical_columns = {}  
    columns = df.columns
    for i in range(len(columns)):
        for j in range(i + 1, len(columns)):
            if df[columns[i]].equals(df[columns[j]]):
                if columns[i] not in identical_columns:
                    identical_columns[columns[i]] = []
                identical_columns[columns[i]].append(columns[j])
    return identical_columns

# Apply the function to DataFrame
identical_cols = find_identical_columns(renamed_col)
print(identical_cols)
```

```python
# Drop identical columns
# List of columns to drop
columns_to_drop = ['PhysicalRecipeID', 'CalcLoopSeq', 'RunStartTime']

# Drop the specified columns
drop_col = renamed_col.drop(columns=columns_to_drop)

# Display the DataFrame to confirm the columns have been removed
display(drop_col.head())
```

### Understanding fluctuation in sensor readings

```python
# Generate the list of runs from 1 to 50
runs_to_plot = list(range(62318, 62348))

# Filter data for the specific runs
filtered_data = drop_col[drop_col['Run'].isin(runs_to_plot)]

# Create a figure with 30 subplots
fig, axs = plt.subplots(10, 3, figsize=(20, 20))  # 10 rows, 3 columns

# Flatten the axes array for easy iteration
axs = axs.flatten()

for i, run in enumerate(runs_to_plot):
    # Filter data for the current run
    run_data = filtered_data[filtered_data['Run'] == run]
    
    # Plotting using the DataFrame index
    axs[i].plot(run_data.index, run_data['Sensor_1'], label=f'Run {run}')
    axs[i].set_title(f'Run {run}')
    axs[i].set_xlabel('Index')
    axs[i].set_ylabel('Sensor Readings')
    axs[i].legend()
    axs[i].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

The graphs do not show a clear indication of anomalous vs non-anomalous runs. 

### Understand correlation between sensor readings

```python
# Select columns of Sensors from 1 to 10
sensor_columns = [f'Sensor_{i}' for i in range(1, 33)]

# Create a DataFrame with the selected columns
selected_data = drop_col[sensor_columns]

# Calculate the correlation matrix
correlation_matrix = selected_data.corr()

# Plot the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix for Sensor_1 to Sensor_33')
plt.show()
```

The correlation matrix shows that there may be some correlation between different sensor readings, but this may be due to the readings being influenced by the same physical environment which can be dictated by parameters such as temperature, pressure etc.
